## Values & Variables

A value or object is a piece of information. The number 42, the text “hello”, the date 27 November 2025, and the dog Fido are all objects. (Python comes with support for numbers, text, and dates, but you’d have to teach it about dogs yourself.) In Python, value and object mean the same thing.

Objects can be made up of other objects. For example: You can put a series of objects in a list. A date is made of a year, month, and day (all numbers). Something called a dictionary lets you associate objects with one other object each, so you can “look up” a “key” (like an English word, which is a text string) and get back the value that you previously associated with that key (like its definition, which is also text string). Fido the dog has a name (text string) and a birthday (date).

Every object in a computer language has a type. The type specifies what you can do with the object, and tells the computer how to do those things. You can add two numbers together, but you can’t add two dogs together. You can ask a dog to bark, but you can’t ask a number to bark.

## Functions
So you now know how to assign a variable, but code is supposed to do stuff right?

Why use functions instead of just directly doing things? Core reasons include you:
- **Can write function once and use it many times** (example: use one function that filters out invalid data on 5 different datasets. If you wrote the logic out 5 times instead of using a function, not only do you have to write a lot more code, but if you change one place without changing all the others you might not have a consistent filter)
- **Can call functions from functions**
- **Code reuse** (maybe you want to use that same filter, but in a different way)
- **Readability**
- **Abstraction: be able to do things without thinking about exactly how they work.** (This is the fundamental skill for managing complexity in larger programs!)

Note:
- `def` keyword (“define” a function)
- Parameters (the variables passed in)
- Indentation
- Descriptive names are better


```python
# Here is how you make a function. This function gets area by multiplying height and width.
# Note it starts with the word def, then the function is named, and then variables, which 
# are called paramters, are passed in between the round brackets. There is a space between 
# def and the name of the function, but no space between the name of the function and ().

def get_area(height,width):
    print(height * width)

# This is the same function, but it's named badly! A function and parameters (the variables
# that are passed into a function) should be named in a way that lets anyone reading the code 
# know what action the function does simply by reading the name. 
def mycoolfunction(x,y):
    print(x*y)

# Call the functions. Note that in the function two parameters are asked for. I pass in two 
# numbers, seperated by a comma. 
get_area(1,5)
mycoolfunction(3,3)

# Don't name your variables like this!
num3=80
num4=38

# name them like this!
door_height=80
door_width=38

# Which is more readable? Can you understand what it does simply by reading it?
mycoolfunction(num3,num4)
get_area(door_height,door_width)
```

```python
# Functions are called (aka executed/run/made to work) by using round brackets, and passing in a parameter.
# Types, such as list, have functions. For example, append adds something to a list. 
print(names)
names.append("bob")
print(names)
```
