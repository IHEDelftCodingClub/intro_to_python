# Control Flow
There are 3 common ways to control flow. A while statement, a in statement, and a for loop. These can be found in all coding languages with slightly different syntax. 

## While statement

````python
# A while statement. A valid way to only print something 10 times. 

# Here I define my function
def print_10_times(mystring):
    count = 0
    while count < 10:
        print(mystring+"x"+str(count))
        count = count+1

# Here I invoke, or call my function.
print_10_times("Puppy")
````

````python
# Here is while statement outside of a function.
count = 0
while count < 5:
    count = count+1
    print(count)
````

## If/else

````python
# Here is an if/else statement.
def is_negative(x):
    if x < 0:
        print("negative")
    elif x == 0:
        print("zero")
    else:
        print("positive")

is_negative(0)
````

````python
# This demonstrates the difference between a print and a return. 
# A return doesn't display the output, but istead sends it back to the 
# place in the code where it was requested. In this case, I am setting the return
# of the function to a variable and then printing it. 

# Often functions are used for "business logic" and return a value, that 
# the requester knows how to display. This gives a division of function between
# logic code and display code that has a variety of advantages. 

def is_negative(x):
    if x < 0:
        return "negative"
    elif x == 0:
        return "zero"
    else:
        return "positive"

the_answer = is_negative(35)
print(the_answer)
````

````python
# You can have as many elif statements as you want
def is_negative(x):
    if x < 0:
        return "negative"
    elif x == 0:
        return "zero"
    elif x == 35:
        return "my age"
    else:
        return "positive"

the_answer = is_negative(35)
print(the_answer)
````

## For loop

````python
animals = ["cat","dog", "fish", "dolphin"]
print(animals)

# This is a for loop
for item in animals:
    print(item, len(item)) # len prints number of char in a string or numbers of items in a list
````

````python
# This for loop shows how to filter a list and save a filtered list
animals = ["cat","dog", "fish", "dolphin"]
animals_that_start_with_d = []
print(animals)

# This is a for loop
for item in animals:
    print(item, len(item))
    if item[0] == 'd':
        print(item+" starts with D!")
        animals_that_start_with_d.append(item)

print(animals_that_start_with_d)
````

# Classes

A class (like Dog) has many **instances** (*individual dogs*)

Some attributes and functions are *shared among all instances* (these are **static**). And you don't need an individual dog to know about or use these attributes or functions. 
- Example: all dogs are mammals

Other attributes and functions are *specific* to one instance 
- Example: each dog has its own name
or can only be done by an individual dog
- Example: an individual dog follows a command

Here, we define a `Dog` class, but all the types, such as int and dict are classes that python has defined. This is why you can say, for example: `my_list.append("my_item")` - `.append()` is a function defined in a `list` class!


````python
# Define a class Dog
class Dog:
    # Init is a special function defined by python. You have to use two underscores (_ _)
    def __init__(self, dog_name):
        self.name = dog_name

    # This function doesn't need an individual Dog to be used - that's what the @staticmethod means
    @staticmethod
    def isMammal():
        return True

    # To access or use this function, you also need an indivudal dog, who has an individual name.
    # Self doesn't have to be passed in if you're calling this function from a dog `mydog.command`
    # as it automatically uses the individual its being called on.
    def command(self, action):
        print(self.name + ", " + action+"!")

# Create two dog instances/individuals
fido = Dog("Fido")
hunter = Dog("Hunter")

# show that they have individual characteristics
print(fido.name)
print(hunter.name)

# Show you can either call a function on the dog, or pass it in
hunter.command("Come")
Dog.command(hunter, "Come")

# Shows you don't need an individual to access the isMammal function
print(Dog.isMammal())
````

## Wait, whats with that weird __init__ thing?

Python has some magic names that it will use for certain purposes without you specifically asking for them. All of these begin and end with two underscores (the character _, which you can type as shift+hyphen). They are called magic methods or dunder methods (dunder is short for double underscore) and `__init__` is pronounced dunder init. When you instantiate a class by calling `Dog(Fido)`, Python creates a new Dog object and then calls `__init__(Fido)` on it. There are lots of these but you wont need most of them. Some examples of more useful ones:
```
	def __eq__(self, other):  # test whether self and other are equal
		# Treat all dogs with the same name as equal to each other.
		# (If you dont define __eq__, a dog is only equal to that specific same object.)
		return isinstance(other, Dog) and self.name == other.name

	def __str__(self):  # return a textual representation of this object
		return <Dog named  + self.name + >
```
